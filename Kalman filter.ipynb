# Kalman filter 

### Model Description

Specifically, $\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{x}_{\mathrm{k}-1}\right)$ is determined by the process noise density $\mathrm{p}\left(\mathbf{w}_{\mathrm{k}}\right)$ with the stateupdate equation:
$$
\mathbf{x}_{\mathrm{k}+1}=\mathbf{F}_{\mathrm{k}+1 . \mathrm{k}} \mathbf{x}_{\mathrm{k}}+\mathbf{w}_{\mathrm{k}} \text { or } \mathbf{x}_{\mathbf{k}+1}=\mathbf{f}\left(\mathrm{k}, \mathbf{x}_{\mathrm{k}}\right)+\mathbf{w}_{\mathrm{k}}
$$
$\mathrm{p}\left(\mathbf{y}_{\mathrm{k}} \mid \mathbf{x}_{\mathrm{k}}\right)$ is determined by the observation noise density $\mathrm{p}\left(\mathbf{v}_{\mathrm{k}}\right)$ with the measurement equation:
$$
\mathbf{y}_{\mathbf{k}}=\mathbf{H}_{\mathbf{k}} \mathbf{x}_{\mathbf{k}}+\mathbf{v}_{\mathbf{k}} \text { or } \mathbf{y}_{\mathbf{k}}=\mathbf{h}\left(\mathrm{k}, \mathbf{x}_{\mathbf{k}}\right)+\mathbf{v}_{\mathbf{k}}
$$
In principle, knowledge of these densities and the initial condition $$ \mathrm{p}\left(\mathbf{x}_{0} \mid \mathbf{y}_{0}\right)$$ determines all $$\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{x}_{\mathrm{k}-1}\right)$$ for all k.

### MMSE
If we set cost function quadratic and the optimal estimate is the mean square error(MMSE) estimator $\hat{x}_{\mathrm{k}}=\mathrm{E}\left(\mathrm{x}_{\mathrm{k}} \mid \mathrm{y}_{1}, \mathrm{y}_{2}, \ldots, \mathrm{y}_{\mathrm{k}}\right)$ where $\mathrm{E}$ is the expctation operator and the cost function $\mathrm{J}_{\mathrm{k}}=$ $\mathrm{E}\left[\tilde{x}_{k}^{2}\right]$
$$
\begin{array}{c}
\mathrm{E}\left[\tilde{\mathbf{x}}_{\mathrm{k}}^{2}\right]=\mathrm{E}\left[\left(\mathbf{x}_{\mathrm{k}}-\hat{\mathbf{x}}_{\mathrm{k}}\right)^{2}\right] \\
=\iint\left(\mathbf{x}_{\mathrm{k}}-\hat{\mathbf{x}}_{\mathbf{k}}\right)^{2} \mathrm{p}\left(\mathbf{x}_{\mathbf{k}}, \mathbf{y}\right) \mathrm{d} \mathbf{x}_{\mathrm{k}} \mathrm{d} \mathbf{y} \\
.=\int\left[\int\left(\mathbf{x}_{\mathrm{k}}-\hat{\mathbf{x}}_{\mathbf{k}}\right)^{2} \mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}\right) \mathrm{d} \mathbf{x}_{\mathrm{k}}\right] \mathrm{p}(\mathbf{y}) \mathrm{d} \mathbf{y}
\end{array}
$$
Now since $\mathrm{p}(\mathbf{y}) \geq 0$ for all $\mathbf{y}$, if we want to minmize the whole integral, we can minimize the interal in brckets for each $\mathbf{y}$. Hence, fixing $\mathbf{y}$ so that we have:
$\frac{\partial}{\vartheta \hat{x}_{\mathrm{k}}} \int\left(\mathbf{x}_{\mathrm{k}}-\hat{\mathbf{x}}_{\mathrm{k}}\right)^{2} \mathrm{p}\left(\mathbf{x}_{\mathbf{k}} \mid \mathbf{y}\right) \mathrm{d} \mathbf{x}_{\mathrm{k}}$
$=\int \frac{\partial}{\vartheta \hat{\mathbf{x}}_{\mathbf{k}}}\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{\mathbf{k}}\right)^{2} \mathrm{p}\left(\mathbf{x}_{\mathbf{k}} \mid \mathbf{y}\right) \mathrm{d} \mathbf{x}_{\mathbf{k}}$
$=\int-2\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{\mathbf{k}}\right) \mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}\right) \mathrm{d} \mathbf{x}_{\mathrm{k}}$
$=-2 \int \mathbf{x}_{\mathrm{k}} \mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}\right) \mathrm{d} \mathbf{x}_{\mathbf{k}}+2 \hat{\mathbf{x}}_{\mathrm{k}} \int \mathbf{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}\right) \mathrm{d} \mathbf{x}_{\mathbf{k}}$
which when set equal to zeros results in
$$
\hat{\mathbf{x}}_{\mathrm{k}}=\int \mathbf{x}_{\mathrm{k}} \mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}\right) \mathrm{d} \mathbf{x}_{\mathrm{k}}
$$
because we have $\int p\left(\boldsymbol{x}_{k} \mid \boldsymbol{y}\right) d \boldsymbol{x}_{k}=1, p\left(\boldsymbol{x}_{k} \mid \boldsymbol{y}\right)$ is a conditional probability density function thus its integral is 1 .
finally we have the optimal estimator : the conditional mean

### Optimal recursive estimation
Given observations $\mathbf{y}_{\mathrm{k}}$, the goal is to estimate the state $\mathbf{x}_{\mathrm{k}}$. with no assumptions about the of the system dynamics at this point. The optimal estimate in the minimum mean-squared error (MMSE) sense is given by the conditional mean which we have proved following :
$$
\widehat{\mathbf{x}}_{k}=\mathrm{E}\left[\mathbf{x}_{\mathbf{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}}\right]
$$
To compute the expectation requires know the posterior density $\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}}\right)$
$$
\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}}\right)=\frac{\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathrm{y}_{2}, \ldots, y_{\mathrm{k}-1}\right) \mathrm{p}\left(\mathrm{y}_{\mathrm{k}} \mid \mathbf{x}_{\mathrm{k}}\right)}{\mathrm{p}\left(y_{\mathrm{k}} \mid y_{1}, \mathrm{y}_{2}, \ldots, \mathrm{y}_{\mathrm{k}-1}\right)}
$$
where
$$
\begin{array}{l}
\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right)=\int \mathrm{p}\left(\mathbf{x}_{\mathrm{k}}, \mathbf{x}_{\mathrm{k}-1} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right) \mathrm{d} \mathbf{x}_{\mathrm{k}-1} \\
=\int \mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{x}_{\mathrm{k}-1}, \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathbf{k}-1}\right) \mathrm{p}\left(\mathbf{x}_{\mathrm{k}-1} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right) \mathrm{d} \mathbf{x}_{\mathbf{k}-1} \\
\quad \mathrm{p}\left(\mathbf{y}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right)=\int \mathrm{p}\left(\mathbf{x}_{\mathrm{k}}, \mathbf{y}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right) \mathrm{d} \mathbf{x}_{\mathbf{k}} \\
\quad=\int \mathrm{p}\left(\mathbf{y}_{\mathrm{k}} \mid \mathbf{x}_{\mathrm{k}}, \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right) \mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right) \mathrm{d} \mathbf{x}_{\mathrm{k}}
\end{array}
$$
Compare the p $\left(\mathbf{x}_{\mathrm{k}-1} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right)$ in the first integral with $\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}}\right)$ which we want to compute, we find it is the posterior density at time $\mathrm{k}-1$, thus we have a recursive equation. And $p\left(\mathbf{y}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right)$ is a normalizing constant, this recursion specifies the current state density as a function of the previous density and the most recent measurement data. The optimal method to recursively update the posterior density as new observations arrive is given by the recursive Bayesian estimation algorithm above. $\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{x}_{\mathrm{k}-1}, \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right)$ and $\mathrm{p}\left(\mathbf{y}_{\mathrm{k}} \mid \mathbf{x}_{\mathrm{k}}, \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right)$ in the integrals can be specified and simplified by the state-space model.

### Consider the Dynamics
$$
\mathbf{x}_{\mathrm{k}+1}=\mathbf{F}_{\mathrm{k}+1 . \mathrm{k}} \mathbf{x}_{\mathrm{k}}+\mathbf{w}_{\mathrm{k}} \text { or } \mathbf{x}_{\mathbf{k}+1}=\mathbf{f}\left(\mathrm{k}, \mathbf{x}_{\mathrm{k}}\right)+\mathbf{w}_{\mathrm{k}}
$$
where $\mathbf{F}_{\mathrm{k}+1, \mathrm{k}}$ is the transition matrix explaining how the state $\mathbf{x}_{\mathrm{k}}$ change from time $\mathrm{k}$ to time $\mathrm{k}+1$. $\mathbf{y}_{\mathrm{k}}$ is the observable at time $\mathrm{k}$ and $\mathbf{H}_{\mathrm{k}}$ is the measurement matrix. $\mathbf{w}_{\mathrm{k}}$ and $\mathbf{v}_{\mathrm{k}}$ are independent, zero-mean, Gaussian noise processes with covariance matrices $\mathbf{Q}_{\mathrm{k}}$ and $\mathbf{R}_{\mathrm{k}}$.
Initialization: For $\mathrm{k}=0$, set
$$
\begin{array}{c}
\widehat{\boldsymbol{x}}_{0}=\mathrm{E}\left[\mathbf{x}_{0}\right] \\
\mathbf{P}_{0}=\mathrm{E}\left[\left(\mathbf{x}_{0}-\mathrm{E}\left[\widehat{\boldsymbol{x}}_{0}\right]\right)\left(\mathbf{x}_{0}-\mathrm{E}\left[\widehat{\boldsymbol{x}}_{0}\right]\right)^{\mathrm{T}}\right]
\end{array}
$$
This choice for the initial conditions not only is intuitively satisfying but also has the advantage of yielding an unbiased estimate of the state $x_{k}$.
Computation: For $\mathrm{k}=1,2, \ldots$, compute:
State estimate propagation
$$
\hat{x}_{k}^{-}=\mathbf{F}_{\mathrm{k}, \mathrm{k}-1} \hat{x}_{k-1}^{-}
$$
Error covariance propagation
$$
\widehat{\mathbf{P}}_{\mathrm{k}}^{-}=\mathbf{F}_{\mathrm{k}, \mathrm{k}-1} \mathbf{P}_{\mathrm{k}-1} \mathbf{F}_{\mathrm{k}, \mathrm{k}-1}^{\mathrm{T}}+\mathbf{Q}_{\mathrm{k}-1}
$$
Kalman gain matrix
$$
\mathbf{G}_{\mathbf{k}}=\mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{\mathrm{k}}^{\mathrm{T}}\left[\mathbf{H}_{\mathbf{k}} \mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{\mathbf{k}}^{\mathrm{T}}+\mathbf{R}_{\mathbf{k}}\right]^{-1}
$$
State estimate update
$$
\widehat{x}_{\mathrm{k}}=\hat{x}_{k}^{-}+\mathbf{G}_{\mathrm{k}}\left(\mathbf{y}_{\mathrm{k}}-\mathbf{H}_{\mathrm{k}} \hat{x}_{k}^{-}\right)
$$
Error covariance update
$$
\mathbf{P}_{\mathrm{k}}=\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}} \mathbf{P}_{\mathrm{k}}^{-}\right)
$$

### Geometric interpretation



### Simplify by Gaussian distribution
A random variable $\boldsymbol{x} \in \mathbb{R}^{\boldsymbol{n}}$ has a Gaussian distribution with $\boldsymbol{m} \in \mathbb{R}^{n}$ and covariance $\boldsymbol{P} \in$ $\mathbb{R}^{n}$ with probability density function:
$$
\mathrm{N}(\mathbf{x} \mid \mathbf{m}, \mathbf{P})=\frac{1}{(\sqrt{2 \pi})^{\frac{N}{2}}|\mathbf{P}|^{\frac{1}{2}}} \exp \left(-(\mathbf{x}-\mathbf{m})^{\mathrm{T}} \mathbf{P}^{-1}(\mathbf{x}-\mathbf{m})\right)
$$
Where $|\mathbf{P}|$ is the determinant of the matrix $\mathbf{P}$.

If we add the assumption that the process noise $\mathbf{w}_{k}$ and measurement noise $\mathbf{v}_{k}$ are both Gaussion white noise with pdf $\mathrm{N}\left(0, \mathbf{R}_{\mathrm{k}}\right)$ and $\mathrm{N}\left(0, \mathbf{Q}_{\mathrm{k}}\right)$, we can model:
$$
\begin{array}{c}
\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right)=N\left(\mathbf{x}_{\mathrm{k}} \mid \mathrm{m}_{\mathrm{k}}^{-}, \mathrm{P}_{\mathrm{k}}^{-}\right) \\
\mathrm{p}\left(\mathbf{x}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}}\right)=N\left(\mathbf{x}_{\mathrm{k}} \mid \mathrm{m}_{\mathrm{k}}, \mathrm{P}_{\mathrm{k}}\right) \\
\mathrm{p}\left(\mathbf{y}_{\mathrm{k}} \mid \mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}\right)=N\left(\mathbf{y}_{\mathrm{k}} \mid \mathrm{H}_{\mathrm{k}} \mathrm{m}_{\mathrm{k}}^{-}, \mathrm{S}_{\mathrm{k}}\right)
\end{array}
$$
The perdiction step is
$$
\begin{array}{c}
\mathrm{m}_{\mathrm{k}}^{-}=\mathbf{F}_{\mathrm{k} . \mathrm{k}-1} \mathrm{~m}_{\mathrm{k}-1} \\
\mathrm{P}_{\mathrm{k}}^{-}=\mathbf{F}_{\mathrm{k} \cdot \mathrm{k}-1} \mathrm{P}_{\mathrm{k}-1} \mathbf{F}_{\mathrm{k} \cdot \mathrm{k}-1}{ }^{T}+\mathbf{Q}_{\mathrm{k}-1}
\end{array}
$$
The update step is
$$
\begin{array}{c}
\tilde{\mathbf{y}}_{\mathbf{k}}=\mathbf{y}_{\mathbf{k}}-\mathbf{H}_{\mathrm{k}} \mathbf{m}_{\mathbf{k}}^{-} \\
\mathbf{G}_{\mathbf{k}}=\mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{k}^{T}\left[\mathbf{H}_{\mathbf{k}} \mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{k}^{T}+\mathbf{R}_{\mathbf{k}}\right]^{-1}
\end{array}
$$

$$
\begin{array}{c}
\mathbf{m}_{\mathrm{k}}=\mathbf{m}_{\mathrm{k}}^{-}+\mathbf{G}_{\mathrm{k}} \tilde{\mathbf{y}}_{\mathbf{k}} \\
\mathbf{P}_{\mathrm{k}}=\mathbf{P}_{\mathbf{k}}^{-}-\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathbf{k}}\right) \mathbf{P}_{\mathbf{k}}^{-}
\end{array}
$$
The joint distribution of $\mathbf{x}_{\mathrm{k}}$ and $\mathbf{x}_{\mathrm{k}-1}$ given $\mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}$ and the marginal distribution of $\mathbf{x}_{\mathrm{k}}$ is compute by equation in the 2.1.3; Similarly, joint distribution of $\mathbf{x}_{\mathrm{k}}$ and $\mathbf{y}_{\mathrm{k}}$ given $\mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}$ and the marginal distribution of $\mathbf{y}_{\mathrm{k}}$ is compute by equation in the $2.1 .3 .$ For the reason that Gaussian distribution is compeleted decided by its mean and variance, the whole process just need to propagate the mean and variance.

### Inital Proof
Suppose we have a measurement on a linear dynamic systerm described by linear equation at time $\mathrm{k}$, we have measurement $\mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1}$, we need to use the new measurement $\mathbf{y}_{\mathrm{k}}$ to update the estimate of the unknow state $\mathbf{x}_{\mathrm{k}}$. As the estimator is a linear function of data we have. We express the estimate $\hat{x}_{k}$ as a combination of the prior estimate $\hat{\mathbf{x}}_{\mathbf{k}}^{-}$and the new measurement $\mathbf{y}_{\mathbf{k}}$ :
$$
\widehat{\boldsymbol{x}}_{k}=\mathbf{G}_{\mathbf{k}}^{\prime} \hat{\mathbf{x}}_{\mathbf{k}}^{-}+\mathbf{G}_{\mathrm{k}} \mathbf{y}_{\mathbf{k}}
$$
What we have to do is find the coefficient matrix $\boldsymbol{G}_{k}$ and $\boldsymbol{G}_{k}^{\prime}$ of the optimal solution at every time $\mathrm{k}$
The state-error vector is:
$$
\widetilde{\boldsymbol{x}}_{\mathrm{k}}=\mathbf{x}_{\mathrm{k}}-\widehat{\boldsymbol{x}}_{\mathrm{k}}
$$
Notice that
$$
\mathrm{E}\left[\tilde{\mathbf{x}}_{\mathbf{k}} \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right]=\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\widehat{\boldsymbol{x}}_{\mathrm{k}}\right) \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right]=\mathbf{0} \quad \text { for all } \mathrm{i}=1,2, \ldots, \mathrm{k}-1
$$
we get:
$$
\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\widehat{\boldsymbol{x}}_{\mathbf{k}}\right) \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right]=\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\mathbf{G}_{\mathbf{k}}^{\prime} \hat{\mathbf{x}}_{\mathbf{k}}^{-}+\mathbf{G}_{\mathrm{k}} \boldsymbol{y}_{\mathbf{k}}\right) \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right]
$$
Then
$$
\mathrm{E}\left[\left(\mathrm{x}_{\mathrm{k}}-\mathbf{G}_{\mathrm{k}}^{\prime} \hat{\mathbf{x}}_{\mathrm{k}}^{-}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}} \mathbf{x}_{\mathrm{k}}-\mathbf{G}_{\mathrm{k}} \mathbf{w}_{k}\right) \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right]=0 \text { for all } \mathrm{i}=1,2, \ldots, \mathrm{k}-1
$$
Since the process noise $\mathbf{w}_{k}$ and measurement noise $\mathbf{v}_{k}$ are uncorrelated, it follows that
$$
\mathrm{E}\left[\mathbf{w}_{k} \mathbf{y}_{\mathbf{i}}^{\mathrm{T}}\right]=\mathbf{0}
$$
it can rewrite as
$$
\begin{array}{c}
\mathrm{E}\left[\left(\mathbf{x}_{\mathrm{k}}-\mathbf{G}_{\mathrm{k}}^{\prime} \hat{\mathbf{x}}_{\mathbf{k}}^{-}-\mathbf{G}_{\mathrm{k}}^{\prime} \mathbf{x}_{\mathbf{k}}+\mathbf{G}_{\mathbf{k}}^{\prime} \mathbf{x}_{\mathbf{k}}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathbf{k}} \mathbf{x}_{\mathbf{k}}-\mathbf{G}_{\mathbf{k}} \mathbf{w}_{k}\right) \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right] \\
=\mathrm{E}\left[\left(\mathbf{I}-\mathbf{G}_{\mathbf{k}}^{\prime}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathbf{k}}\right) \mathbf{x}_{\mathbf{k}} \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}+\mathbf{G}_{\mathbf{k}}^{\prime}\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{\mathbf{k}}^{-}\right) \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right]=0
\end{array}
$$
where I is the identity matrix
it can simplifies to
$$
\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}}^{\prime}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}}\right) \mathrm{E}\left[\mathbf{x}_{\mathrm{k}} \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right]=\mathbf{0} \text { for all } \mathrm{i}=1,2, \ldots, \mathrm{k}-1
$$
with the principle of orthogonality
$$
\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\widehat{\mathbf{x}}_{\mathbf{k}}^{-}\right) \mathbf{y}_{\mathrm{i}}^{\mathrm{T}}\right]=\mathbf{0} \quad \text { for all } \mathrm{i}=1,2, \ldots, \mathrm{k}-1
$$
So for any state $\mathbf{x}_{\mathrm{k}}$ and observable $\mathbf{y}_{\mathrm{i}},(2-7)$ can only be satisfied if :
$$
\mathbf{I}-\mathbf{G}_{\mathrm{k}}^{\prime}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}}=\mathbf{0}
$$
Also can be written as :
$$
\mathbf{G}_{\mathrm{k}}^{\prime}=\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}}
$$
Substituting, we can get a posterior estimate
$$
\begin{aligned}
\widehat{\boldsymbol{x}}_{k} &=\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}}\right) \hat{\mathbf{x}}_{\mathbf{k}}^{-}+\mathbf{G}_{\mathrm{k}} \mathbf{y}_{\mathbf{k}} \\
=& \hat{\mathbf{x}}_{\mathbf{k}}^{-}+\mathbf{G}_{\mathrm{k}}\left(\mathbf{y}_{\mathbf{k}}-\mathbf{H}_{\mathrm{k}} \hat{\mathbf{x}}_{\mathbf{k}}^{-}\right)
\end{aligned}
$$
The matrix $\mathbf{G}_{\mathrm{k}}$ is what we need to figure out, which is called the Kalman gain.
Next, we derive an explict formula for $\mathbf{G}_{\mathrm{k}}^{\prime}$, from the principle of orthogonality, we have
$$
\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k}\right) \mathbf{y}_{\mathrm{k}}^{\mathrm{T}}\right]=\mathbf{0}
$$
It follows that
$$
\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k}\right) \hat{\boldsymbol{y}}_{\mathbf{k}}^{\mathrm{T}}\right]=\mathbf{0}
$$
Where $\widehat{\boldsymbol{y}}_{\mathrm{k}}$ is an estimate of $\mathbf{y}_{\mathrm{k}}$ give the previous measurement $\mathbf{y}_{1}, \mathbf{y}_{2}, \ldots, \mathbf{y}_{\mathrm{k}-1} .$ Define the innovations process
$$
\tilde{\mathbf{y}}_{\mathrm{k}}=\mathbf{y}_{\mathrm{k}}-\hat{\mathbf{y}}_{\mathrm{k}}
$$
The innovation process represents a measure of the "new" information contained in $\mathbf{y}_{\mathrm{k}}$; it may also be expressed as
$$
\begin{aligned}
\tilde{\mathbf{y}}_{\mathbf{k}} &=\mathbf{y}_{\mathrm{k}}-\mathbf{H}_{\mathrm{k}} \hat{\mathbf{x}}_{\mathbf{k}}^{-} \\
&=\mathbf{H}_{\mathbf{k}} \mathbf{x}_{\mathbf{k}}+\mathbf{v}_{\mathbf{k}}-\mathbf{H}_{\mathbf{k}} \hat{\mathbf{x}_{\mathbf{k}}}^{-} \\
&=\mathbf{H}_{\mathbf{k}}\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{\mathbf{k}}^{-}\right)+\mathbf{v}_{\mathbf{k}} \\
&=\mathbf{H}_{\mathbf{k}} \tilde{\boldsymbol{x}}_{\mathbf{k}}^{-}+\mathbf{v}_{\mathbf{k}}
\end{aligned}
$$
subtract
$$
\begin{array}{c}
\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k}\right) \mathbf{y}_{\mathrm{k}}^{\mathrm{T}}\right]-\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k}\right) \widehat{\boldsymbol{y}}_{\mathbf{k}}^{\mathrm{T}}\right]=\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k}\right)\left(\mathbf{y}_{\mathrm{k}}^{\mathrm{T}}-\widehat{\boldsymbol{y}}_{\mathbf{k}}^{\mathrm{T}}\right)\right] \\
=\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k}\right) \tilde{\mathbf{y}}_{\mathbf{k}}^{\mathrm{T}}\right]=\mathbf{0}
\end{array}
$$
we rewrite the state error $\mathbf{x}_{\mathrm{k}}-\hat{\mathbf{x}}_{k}$
$$
\begin{aligned}
\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k} &=\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{\mathbf{k}}^{-}-\mathbf{G}_{\mathrm{k}}\left(\mathbf{y}_{\mathbf{k}}-\mathbf{H}_{\mathbf{k}} \hat{\mathbf{x}}_{\mathbf{k}}^{-}\right) \\
&=\tilde{\mathbf{x}}_{\mathbf{k}}^{-}-\mathbf{G}_{\mathrm{k}}\left(\mathbf{H}_{\mathbf{k}} \mathbf{x}_{\mathbf{k}}+\mathbf{v}_{\mathbf{k}}-\mathbf{H}_{\mathrm{k}} \hat{\mathbf{x}}_{\mathbf{k}}^{-}\right) \\
&=\tilde{\mathbf{x}}_{\mathbf{k}}^{-}-\mathbf{G}_{\mathrm{k}}\left(\mathbf{H}_{\mathbf{k}} \tilde{\mathbf{x}}_{\mathbf{k}}^{-}+\mathbf{v}_{\mathbf{k}}\right) \\
&=\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathbf{k}}\right) \tilde{\mathbf{x}}_{\mathbf{k}}^{-}-\mathbf{G}_{\mathrm{k}} \mathbf{v}_{\mathbf{k}}
\end{aligned}
$$
Hence, subtituting 
$$
\mathrm{E}\left[\left\{\left(\left(\mathbf{I}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathrm{k}}\right) \tilde{\mathbf{x}}_{\mathbf{k}}^{-}-\mathbf{G}_{\mathrm{k}} \mathbf{v}_{\mathbf{k}}\right)\left(\mathbf{H}_{\mathbf{k}} \tilde{\mathbf{x}}^{-}+\mathbf{v}_{\mathbf{k}}\right)^{T}\right\}\right]=\mathbf{0}
$$
Since $\mathbf{v}_{\mathrm{k}}$ is independent of the state $\mathbf{x}_{\mathrm{k}}$ and therefore the error $\widetilde{\boldsymbol{x}}_{\mathrm{k}}^{-}$, so it reduce to
$$
\left(\mathbf{I}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathbf{k}}\right) \mathrm{E}\left[\tilde{\boldsymbol{x}}_{\mathbf{k}}^{-} \widetilde{\boldsymbol{x}}_{\mathbf{k}}^{\mathrm{T}-}\right] \mathbf{H}_{k}^{T}-\mathbf{G}_{\mathbf{k}} \mathrm{E}\left[\mathbf{v}_{\mathbf{k}} \mathbf{v}_{k}^{T}\right]=\mathbf{0}
$$
Now we define a prior covariance matrix
$$
\mathbf{P}_{\mathrm{k}}^{-}=\mathrm{E}\left[\left(\mathbf{x}_{\mathrm{k}}-\hat{\mathbf{x}}_{\mathrm{k}}^{-}\right)\left(\mathbf{x}_{\mathrm{k}}-\hat{\mathbf{x}}_{\mathrm{k}}^{-}\right)^{\mathrm{T}}\right]=\mathrm{E}\left[\tilde{\boldsymbol{x}}_{\mathrm{k}}^{-} \tilde{\boldsymbol{x}}_{\mathrm{k}}^{\mathrm{T}-}\right]
$$
Rewrite 
$$
\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}}\right) \mathbf{P}_{\mathrm{k}}^{-} \mathbf{H}_{k}^{T}-\mathbf{G}_{\mathrm{k}} \mathbf{R}_{\mathbf{k}}=\mathbf{0}
$$
To get $\mathbf{G}_{\mathrm{k}}$, we get the desired formula

$$
\mathbf{G}_{\mathrm{k}}=\mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{k}^{T}\left[\mathbf{H}_{\mathrm{k}} \mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{k}^{T}+\mathbf{R}_{\mathrm{k}}\right]^{-1}
$$
where the symbol $[*]^{-1}$ denotes the inverse of the matrix inside the square brackets. Equation is the desired formula for computing the Kalman gain $\mathbf{G}_{\mathrm{k}}$, which is defined in terms of the a priori covariance matrix $\mathbf{P}_{\mathrm{k}}^{-}$.
Now we find rules to update the posterior convariance matrix $\mathbf{P}_{\mathrm{k}}$ at time $\mathrm{k}$, when $\mathbf{P}_{\mathrm{k}}^{-}$is computed by 
$$
\mathbf{P}_{\mathbf{k}}=\mathrm{E}\left[\tilde{\mathbf{x}}_{\mathbf{k}} \tilde{\mathbf{x}}_{\mathrm{k}}^{\mathrm{T}}\right]=\mathrm{E}\left[\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k}\right)\left(\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{k}\right)^{T}\right]
$$
we substitute 
$$\left.\left.\mathbf{P}_{\mathbf{k}}=\mathrm{E}\left[\left\{\left(\mathbf{I}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathbf{k}}\right) \widetilde{\boldsymbol{x}}_{\mathbf{k}}^{-}-\mathbf{G}_{\mathrm{k}} \mathbf{v}_{\mathbf{k}}\right)\right)\left(\left(\mathbf{I}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathrm{k}}\right) \widetilde{\boldsymbol{x}}_{\mathrm{k}}^{-}-\mathbf{G}_{\mathrm{k}} \mathbf{v}_{\mathbf{k}}\right)^{T}\right\}\right]$$
$$=\left(\mathbf{I}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathbf{k}}\right) \mathrm{E}\left[\widetilde{\boldsymbol{x}}_{\mathbf{k}}^{-} \widetilde{\boldsymbol{x}}_{\mathbf{k}}^{\mathrm{T}}\right]\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathbf{k}}\right)^{T}+\mathbf{G}_{\mathrm{k}} \mathbf{R}_{\mathbf{k}} \mathbf{G}_{k}^{T}$$
$$=\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}}\right) \mathbf{P}_{\mathrm{k}}^{-}\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}}\right)^{T}+\mathbf{G}_{\mathbf{k}} \mathbf{R}_{\mathbf{k}} \mathbf{G}_{1}^{T}$$
Now we derivate a simplified form by expanding terms:
$$
\begin{array}{c}
\mathbf{P}_{\mathbf{k}}=\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathbf{k}}\right) \mathbf{P}_{\mathbf{k}}^{-}-\left(\mathbf{I}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathbf{k}}\right) \mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{k}^{T} \mathbf{G}_{k}^{T}+\mathbf{G}_{\mathbf{k}} \mathbf{R}_{\mathbf{k}} \mathbf{G}_{k}^{T} \\
=\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathbf{k}}\right) \mathbf{P}_{\mathbf{k}}^{-}-\mathbf{G}_{\mathbf{k}} \mathbf{R}_{\mathrm{k}} \mathbf{G}_{k}^{T}+\mathbf{G}_{\mathbf{k}} \mathbf{R}_{\mathbf{k}} \mathbf{G}_{k}^{T} \\
=\left(\mathbf{I}-\mathbf{G}_{\mathrm{k}} \mathbf{H}_{\mathrm{k}}\right) \mathbf{P}_{\mathbf{k}}^{-}
\end{array}
$$
where we use simplified equation and right multiplication a term $\mathbf{G}_{k}^{T}$
$$
\left(\mathbf{I}-\mathbf{G}_{\mathbf{k}} \mathbf{H}_{\mathbf{k}}\right) \mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{k}^{T} \mathbf{G}_{k}^{T}-\mathbf{G}_{\mathrm{k}} \mathbf{R}_{\mathbf{k}} \mathbf{G}_{k}^{T}=\mathbf{0}
$$
So we need to compute $\mathbf{P}_{\mathrm{k}}^{-}$to get $\mathbf{P}_{\mathrm{k}}$, we first recognize that the a priori estimate of the state is defined in terms of the "old' a posteriori estimate as follows:
$$
\begin{array}{c}
\hat{\mathbf{x}}_{\mathbf{k}}^{-}=\mathbf{F}_{\mathbf{k}, \mathbf{k}-1} \hat{\mathbf{x}}_{k-1} \\
\widetilde{\boldsymbol{x}}_{\mathbf{k}}^{-}=\mathbf{x}_{\mathbf{k}}-\hat{\mathbf{x}}_{\mathbf{k}}^{-}=\mathbf{F}_{\mathbf{k} . \mathbf{k}-1} \mathbf{x}_{\mathbf{k}-1}+\mathbf{w}_{\mathbf{k}-1}-\mathbf{F}_{\mathbf{k} . \mathbf{k}-1} \hat{\mathbf{x}}_{k-1} \\
=\mathbf{F}_{\mathrm{k} . \mathrm{k}-1}\left(\mathbf{x}_{\mathbf{k}-1}-\hat{\mathbf{x}}_{k-1}\right)+\mathbf{w}_{\mathbf{k}-1} \\
=\mathbf{F}_{\mathbf{k} . \mathbf{k}-1} \tilde{\mathbf{x}}_{\mathbf{k}-1}+\mathbf{w}_{\mathbf{k}-1}
\end{array}
$$
$\mathbf{P}_{\mathbf{k}}^{-}=\mathrm{E}\left[\widetilde{\boldsymbol{x}}_{\mathrm{k}}^{-} \tilde{\boldsymbol{x}}_{\mathrm{k}}^{\mathrm{T}-}\right]$
$=\mathbf{F}_{\mathbf{k} . \mathbf{k}-\mathbf{1}} \mathrm{E}\left[\tilde{\mathbf{x}}_{\mathbf{k}-1} \tilde{\mathbf{x}}_{k-1}^{T}\right] \mathbf{F}_{\mathbf{k} . \mathbf{k}-1}{ }^{T}+\mathrm{E}\left[\mathbf{w}_{\mathbf{k}-1} \mathbf{w}_{\mathbf{k}-\mathbf{1}}^{T}\right]$
$=\mathbf{F}_{\mathbf{k}, \mathbf{k}-\mathbf{1}} \mathbf{P}_{\mathbf{k}-1} \mathbf{F}_{\mathbf{k} \cdot \mathbf{k}-1}{ }^{T}+\mathbf{Q}_{\mathbf{k}-1}$
which defines the dependence of the a priori covariance matrix $\mathbf{P}_{\mathrm{k}}^{-}$on the "old" a posteriori covariance matrix $\mathbf{P}_{\mathrm{k}-1}$


### Case study from https://machinelearningspace.com/2d-object-tracking-using-kalman-filter/
2-D Object Tracking using Kalman Filter in Python  
simple kinematic equation $y=x_0+v_0*t+1/2*a*t^2$  
$$
\mathbf{x}_{k}=\left[\begin{array}{c}
x_{k} \\
y_{k} \\
\dot{x}_{k} \\
\dot{y}_{k}
\end{array}\right]=\left[\begin{array}{c}
x_{k-1}+\dot{x}_{k-1} \Delta t+1 / 2 \ddot{x}_{k-1} \Delta t^{2} \\
y_{k-1}+\dot{y}_{k-1} \Delta t+1 / 2 \ddot{y}_{k-1} \Delta t^{2} \\
\dot{x}_{k-1}+\ddot{x}_{k-1} \Delta t \\
\dot{y}_{k-1}+\ddot{y}_{k-1} \Delta t
\end{array}\right]
$$
write it into the form of matrix multiplication as follows:
$$
\mathbf{x}_{k}=\left[\begin{array}{c}
x_{k} \\
y_{k} \\
\dot{x}_{k} \\
\dot{y}_{k}
\end{array}\right]=\left[\begin{array}{cccc}
1 & 0 & \Delta t & 0 \\
0 & 1 & 0 & \Delta t \\
0 & 0 & 1 & 0 \\
0 & 0 & 0 & 1
\end{array}\right]\left[\begin{array}{l}
x_{k-1} \\
y_{k-1} \\
\dot{x}_{k-1} \\
\dot{y}_{k-1}
\end{array}\right]+\left[\begin{array}{cc}
\frac{1}{2}(\Delta t)^{2} & 0 \\
0 & \frac{1}{2}(\Delta t)^{2} \\
\Delta t & 0 \\
0 & \Delta t
\end{array}\right]\left[\begin{array}{l}
\ddot{x}_{k-1} \\
\ddot{y}_{k-1}
\end{array}\right]
$$
assume the linear full state observation z=Hx+v. 
$$
\mathbf{z}_{k}=\left[\begin{array}{llll}
1 & 0 & 0 & 0 \\
0 & 1 & 0 & 0
\end{array}\right]\left[\begin{array}{c}
x_{k} \\
y_{k} \\
\dot{x}_{k} \\
\dot{y}_{k}
\end{array}\right]+\mathbf{v}_{k}
$$
transformation matrix $H$ as:
$$
\mathbf{H}=\left[\begin{array}{llll}
1 & 0 & 0 & 0 \\
0 & 1 & 0 & 0
\end{array}\right]
$$
Design covariance noise $Q$(https://machinelearningspace.com/object-tracking-python/#noise-covariance)
$$
\mathbf{Q}=\left[\begin{array}{cccc}
\frac{\Delta t^{4}}{4} & 0 & \frac{\Delta t^{3}}{2} & 0 \\
0 & \frac{\Delta t^{4}}{4} & 0 & \frac{\Delta t^{3}}{2} \\
\frac{\Delta t^{3}}{2} & 0 & \Delta t^{2} & 0 \\
0 & \frac{\Delta t^{3}}{2} & 0 & \Delta t^{2}
\end{array}\right] \sigma_{a}^{2}
$$
Design measurement noise $R$, x and y is independent, just use $\mathbf{I}$

```python
## state[p v]^T
import numpy as np
class KalmanFilter(object):
    def __init__(self,x,u,std_a,std_x,std_y,dt):
        self.dt=dt
        self.x=x
        self.u=u
        self.A = np.matrix([[1, 0, self.dt, 0],
                            [0, 1, 0, self.dt],
                            [0, 0, 1, 0],
                            [0, 0, 0, 1]])
        self.B = np.matrix([[(self.dt**2)/2, 0],
                            [0, (self.dt**2)/2],
                            [self.dt,0],
                            [0,self.dt]])
        self.H = np.matrix([[1, 0, 0, 0],
                            [0, 1, 0, 0]])
        self.Q = np.matrix([[(self.dt**4)/4, 0, (self.dt**3)/2, 0],
                            [0, (self.dt**4)/4, 0, (self.dt**3)/2],
                            [(self.dt**3)/2, 0, self.dt**2, 0],
                            [0, (self.dt**3)/2, 0, self.dt**2]]) * std_a**2
        self.R = np.matrix([[std_x**2,0],
                           [0, std_y**2]])
        
    
        self.P = np.eye(self.A.shape[1])
    def predict(self):
       ### get x_k+1^{-} just by systerm dynamics wait for measure
        
        self.x=self.A*self.x+self.B*self.u
        
        self.P=self.A*self.P*self.A.T+self.Q
        return self.x[0:2]
        
        
    def update(self,y):
        
        """
        compute the kalman gain
        by
        $$
        \mathbf{G}_{\mathrm{k}}=\mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{k}^{T}\left[\mathbf{H}_{\mathrm{k}} \mathbf{P}_{\mathbf{k}}^{-} \mathbf{H}_{k}^{T}+\mathbf{R}_{\mathrm{k}}\right]^{-1}
        $$
        """
        D=self.H*self.P*self.H.T+self.R
        G=self.P*self.H.T*np.linalg.inv(D)
        self.x=self.x+G*(y-self.H*self.x)
        I=np.eye(self.H.shape[1])
        self.P=self.P-(I-G*self.H)*self.P
        return self.x[0:2]
```

```python
"""
Design An simple object detector.
no marching learning just classical image processing algorithms.
convert a video frame passed through the frame argument to a Grayscale image 
using cv2.cvtColor(). 
After that, it detects the edge of the object in the image using Canny edge detection.
"""
import cv2
def detect(frame,debugMode):
    # Convert frame from BGR to GRAY
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    if (debugMode):
        cv2.imshow('gray', gray)

    # Edge detection using Canny function
    img_edges = cv2.Canny(gray,  50, 190, 3)
    if (debugMode):
        cv2.imshow('img_edges', img_edges)

    # Convert to black and white image
    ret, img_thresh = cv2.threshold(img_edges, 254, 255,cv2.THRESH_BINARY)
    if (debugMode):
        cv2.imshow('img_thresh', img_thresh)

    # Find contours
    contours, _ = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Set the accepted minimum & maximum radius of a detected object
    min_radius_thresh= 3
    max_radius_thresh= 30   

    centers=[]
    for c in contours:
        # ref: https://docs.opencv.org/trunk/dd/d49/tutorial_py_contour_features.html
        (x, y), radius = cv2.minEnclosingCircle(c)
        radius = int(radius)

        #Take only the valid circles
        if (radius > min_radius_thresh) and (radius < max_radius_thresh):
            centers.append(np.array([[x], [y]]))
    cv2.imshow('contours', img_thresh)
    return centers
```

```python
## objTracking
def main():

    HiSpeed= 100
    ControlSpeedVar = 30  #Lowest: 1 - Highest:100
    debugMode=1

    #Create KalmanFilter object KF
    #KalmanFilter(x,u,std_a,std_x,std_y,dt):
    x0= np.matrix([[0], [0], [0], [0]])
    u=np.matrix([[1],[1]])
    KF = KalmanFilter(x0, u,1.0, 0.1,0.1,0.1)

    # Create opencv video capture object
    VideoCap = cv2.VideoCapture('video/video_randomball.avi')

    while(True):
        # Read frame
        ret, frame = VideoCap.read()
        
        # Detect object
        centers = detect(frame,debugMode)
        
        # If centroids are detected then track them
        if (len(centers) > 0):

            # Draw the detected circle
            cv2.circle(frame, (int(centers[0][0]), int(centers[0][1])), 10, (0, 191, 255), 2)

            # Predict
            (x, y) = KF.predict()
            # Draw a rectangle as the predicted object position
            cv2.rectangle(frame, (x - 15, y - 15), (x + 15, y + 15), (255, 0, 0), 2)

            # Update
            (x1, y1) = KF.update(centers[0])
            # Draw a rectangle as the estimated object position
            cv2.rectangle(frame, (x1 - 15, y1 - 15), (x1 + 15, y1 + 15), (0, 0, 255), 2)

            cv2.putText(frame, "Estimated Position", (x1 + 15, y1 + 10), 0, 0.5, (0, 0, 255), 2)
            cv2.putText(frame, "Predicted Position", (x + 15, y), 0, 0.5, (255, 0, 0), 2)
            cv2.putText(frame, "Measured Position", (centers[0][0] + 15, centers[0][1] - 15), 0, 0.5, (0,191,255), 2)

        cv2.imshow('image', frame)

        if cv2.waitKey(2) & 0xFF == ord('q'):
            VideoCap.release()
            cv2.destroyAllWindows()
            break

        cv2.waitKey(HiSpeed-ControlSpeedVar+1)


if __name__ == "__main__":
    # execute main
    main()
```
